```python
#Import Library

from google.colab import drive
import cv2
import matplotlib.pyplot as plt

#Mount Google Drive

drive.mount('/content/drive')

#Baca Gambar dari Drive

image_path = '/content/drive/My Drive/Data Penelitian/KlasifikasiHamaSawit/Preprocessing/Data_Original.jpg/Belalang/IMG_8664.jpg'  # Ganti dengan path gambar kamu
original_img = cv2.imread(image_path)
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)  # Konversi BGR ke RGB

#Resize Gambar ke (800, 800)

resize_target = (800, 800)  # (width, height)
resized_img = cv2.resize(original_img, resize_target)

#Crop Gambar dari Hasil Resize

# Format: (left, upper, right, lower)
crop_box = (200, 10, 600, 800)
left, upper, right, lower = crop_box
cropped_img = resized_img[upper:lower, left:right]

#Tampilkan Semua Gambar

plt.figure(figsize=(18, 6))

# Gambar Asli
plt.subplot(1, 3, 1)
plt.imshow(original_img)
plt.title('1. Gambar Asli')
plt.axis('off')

# Gambar Resize
plt.subplot(1, 3, 2)
plt.imshow(resized_img)
plt.title(f'2. Gambar Resize {resize_target}')
plt.axis('off')

# Gambar Crop dari Resize
plt.subplot(1, 3, 3)
plt.imshow(cropped_img)
plt.title(f'3. Crop dari Resize\nBox: {crop_box}')
plt.axis('off')

plt.show()
```
